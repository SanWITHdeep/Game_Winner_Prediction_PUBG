# Data Preprocessing for Modeling

### Import Libraries for Model Training and Prediction

````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
````

Insight:
Importing necessary libraries for model training and evaluation.

### Splitting Data into Features and Target


````python
X = data.drop(columns=['winPlacePerc'])
y = data['winPlacePerc']
````

Insight:
Separating the features (X) and the target variable (y).

### Standardize the Features

````python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
````

Insight:
Standardizing features to have a mean of 0 and a standard deviation of 1.

### Split Data into Training and Testing Sets


````python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
````

Insight:
Splitting the dataset into training (80%) and testing (20%) sets.

# Model Training and Evaluation


### RandomForestRegressor

````python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

mse = mean_squared_error(y_test, y_pred_rf)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_rf)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}")
````

Insight:

Training a RandomForestRegressor model.
Evaluating its performance using Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared metrics.

### LinearRegression

````python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

mse1 = mean_squared_error(y_test, y_pred_lr)
rmse1 = np.sqrt(mse1)
r21 = r2_score(y_test, y_pred_lr)

print(f"Mean Squared Error: {mse1}")
print(f"Root Mean Squared Error: {rmse1}")
print(f"R-squared: {r21}")
````

Insight:

Training a Linear Regression model.
Evaluating its performance using the same metrics as above.

### DecisionTreeRegressor

````python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

mse2 = mean_squared_error(y_test, y_pred_dt)
rmse2 = np.sqrt(mse2)
r22 = r2_score(y_test, y_pred_dt)

print(f"Mean Squared Error: {mse2}")
print(f"Root Mean Squared Error: {rmse2}")
print(f"R-squared: {r22}")
````

Insight:

Training a Decision Tree Regressor model.
Evaluating its performance using the same metrics.
